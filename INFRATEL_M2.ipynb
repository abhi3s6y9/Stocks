```python
df1 = pd.read_csv('Large_Cap/AXISBANK.csv')
df2 = pd.read_csv('Small_Cap/ASHOKA.csv')
df3 = pd.read_csv('Mid_Cap/APOLLOTYRE.csv')
df4 = pd.read_csv('Mid_Cap/BERGEPAINT.csv')
df5 = pd.read_csv('Mid_Cap/PNB.csv')
```

```python
df1.reset_index(drop = True, inplace = True)
df2.reset_index(drop = True, inplace = True)
df3.reset_index(drop = True, inplace = True)
df4.reset_index(drop = True, inplace = True)
df5.reset_index(drop = True, inplace = True)
```

```python
dfN = pd.read_csv('./Nifty50/Nifty50.csv')
dfP = pd.read_csv('./Mid_Cap/PNB.csv')

dfP = dfP[dfP['Series'] == 'EQ']
dfP.reset_index(drop = True, inplace = True)
vnp = pd.concat([dfN['Close'],dfP['Close Price']], axis = 1)

vnp.columns = ['Nifty', 'pnb']
vnp.pct_change()
vnp.dropna(inplace =True,axis =0)
```

```python
points = pd.DataFrame(index = df.index)
points['p'] = np.zeros(points.shape[0])
points['sma21'] = df['Close Price'].rolling(window=21,min_periods=1).mean()
points['sma34'] = df['Close Price'].rolling(window=34,min_periods=1).mean()

fig = plt.figure(figsize = (20,10))
ax1 = fig.add_subplot(111, ylabel='Price in Rupees')
df['Close Price'].plot(ax = ax1, color = 'r')
points[['sma21','sma34']].plot(ax = ax1)
```

```python
points['p'][21:] = np.where(points['sma21'][21:]>points['sma34'][21:],1.0,0.0)
points['position'] = points['p'].diff()

fig = plt.figure(figsize = (20,10))
ax1 = fig.add_subplot(111, ylabel='Price in Rupees')
df['Close Price'].plot(ax = ax1, color = 'r',lw = 2.)
points[['sma21','sma34']].plot(ax = ax1, lw = 1.)
ax1.plot(points.loc[points.position == 1.0].index, points['sma21'][points.position == 1.0],'^',markersize = 10, color = 'g')
ax1.plot(points.loc[points.position == -1.0].index, points['sma21'][points.position == -1.0],'v',markersize = 10, color = 'k')
```

```python
avg = df['Close Price'].rolling(14).mean()
sd = df['Close Price'].rolling(14).std()
upperBand = avg + (sd*2)
lowerBand = avg - (sd*2)

df['avg'] = np.round(avg,3)
df['upper'] = np.round(upperBand,3)
df['lower'] = np.round(lowerBand,3)
```

```python
df['Average Price'].plot(c = 'k', figsize = (20,10), lw = 2)
df['avg'].plot(c='b', figsize = (20,10), lw = 1)
df['upper'].plot(c='g', figsize = (20,10), lw = 1)
df['lower'].plot(c='r', figsize = (20,10), lw = 1)
```
